### Methods for converting calories from a given food into equivalent amount of topsoil (in tons) needed for the same sustenance  

```python
import math
import scipy #skippy
```

Assuming that the numbers in the meme are correct, the number of kCal in a large Oreo chocolate shake and 4.5 metric tons of Loam topsoil are as follows. We can use this to get a conversion factor between kCal consumed and equivalent amount of Loam topsoil needed for the same number of calories.

```python
choccyShake = 2600 #kCal in Chocolate Oreo Shake Large
Loam = 2593 #kCal in 4.5 metric tons of Lowe's Loam topsoil

kCalPerTon = Loam / 4.5
```

```python
#Function which calculates the equivalent loam consumption in tons neccessary for equivalent sustenance
def kCal2Loam(kCal):
    return kCal / kCalPerTon    
```

### Input the amount of calories in your meal below.

```python
meal = 1080 #default is 1080, the number of calories in the entire rotisserie chicken Sam consumed at MDM
```

### Calculate loam equivalent

```python
loamEqui = kCal2Loam(meal)

print("Your boring {} calorie meal is equivalent to {:.2f} metric tons of Loam topsoil \n(which would have been much more interesting and appetizing no doubt). \nCongratulations!".format(meal, loamEqui))
```
