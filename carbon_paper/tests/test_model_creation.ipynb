```python
import surp
from surp.simulation.multizone_sim import create_model
import matplotlib.pyplot as plt
import arya
import numpy as np
```

```python
model = create_model("../temp/", "test", spec="lateburst")
```

```python
t = np.linspace(0, 13, 1000)
Rs = [3, 5, 7, 9, 11, 13]

sfrs = [model.evolution(R, t) for R in Rs]
```

```python
cmap = arya.figure.colorbar.HueMap([0, 15])
for i in range(len(Rs)):
    plt.plot(t, sfrs[i], label=Rs[i], color=cmap(Rs[i]))
    
plt.xlabel("R")
plt.ylabel("sfr")
plt.yscale("log")
plt.ylim(1e-4)
arya.Colorbar(cmap)
```

```python
cmap = arya.figure.colorbar.HueMap([0, 15])
Rs1 = np.linspace(0, 15.5, 1000)

etas = [model.mass_loading(R) for R in Rs1]
plt.plot(Rs1, etas)
plt.xlabel("R")
plt.ylabel(r"$\eta$")
```
