```python
import matplotlib.pyplot as plt
import vice
import arya

import numpy as np
import surp.gce_math as gcem
```

```python
from scipy.integrate import quad
```

```python
imf = vice.imf.kroupa
```

```python
from scipy.interpolate import interp1d
```

```python
def calculate_yield(Ms, ys, Z0, m_low=8, m_high=25):
    y_interp = interp1d(Ms, ys)

    A = quad(lambda M: M * imf(M), 0.08, m_high)[0]
    return quad(lambda M: (y_interp(M) - Z0*(M-1.44)) * imf(M)/A, m_low, m_high)
```

```python
calculate_yield([8, 12, 15, 20, 25], [0, 7.88e-2, 1.553e-1, 2.051e-1, 2.324e-1], 3.4633e-3, m_low=8) # C, Z=0.02
```
