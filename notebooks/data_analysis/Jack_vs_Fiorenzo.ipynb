```python
def compare_stars(x, y, alpha=0.2, s=1, **kwargs):
    plt.scatter(v21[x], v21[y], alpha=alpha, s=s, label="V+21")
    x_r = aah.convert_name(x)
    y_r = aah.convert_name(y)
    plt.scatter(r22[x_r], r22[y_r], alpha=alpha, s=s, label="R+22")
    
    x_l = x.upper().replace("O", "$\\alpha$")
    y_l = y.upper().replace("O", "$\\alpha$")
    plt.xlabel(x_l)
    plt.ylabel(y_l)
```

```python
def compare_means(x, y, alpha=0.2, s=1, xlim=None, **kwargs):
    pluto.plot_mean_track(v21[x], v21[y], label="V+21", xlim=xlim, plot_points=True, err_mean=True)
    x_r = aah.convert_name(x)
    y_r = aah.convert_name(y)
    
    filt = ~np.isnan(r22[x_r])
    filt &= ~np.isnan(r22[y_r])
    pluto.plot_mean_track(r22[filt][x_r], r22[filt][y_r], label="R+22", xlim=xlim, plot_points=True, err_mean=True)
    
    x_l = x.upper().replace("O", "$\\alpha$")
    y_l = y.upper().replace("O", "$\\alpha$")
    plt.xlabel(x_l)
    plt.ylabel(y_l)
    
    pluto.fancy_legend()
```

```python
def compare_means_hi_lo(x, y, alpha=0.2, s=1, xlim=None, **kwargs):
    ha = v21["high_alpha"]
    
    params = {
        "xlim": xlim,
        "plot_points": True,
        "err_mean": False, 
        "min_count": 3,
        "alpha_bars": 0.5
    }
    pluto.plot_mean_track(v21[ha][x], v21[ha][y], label="V+21 high $\\alpha$", 
                          marker="^", c="k", **params)
    pluto.plot_mean_track(v21[~ha][x], v21[~ha][y], label="V+21 low $\\alpha$", 
                          marker="v", c="k", **params)

    x_r = aah.convert_name(x)
    y_r = aah.convert_name(y)
    
    filt = ~np.isnan(r22[x_r])
    filt &= ~np.isnan(r22[y_r])
    
    ha = r22["high_alpha"]
    pluto.plot_mean_track(r22[filt][ha][x_r], r22[filt][ha][y_r], label="subgiants high $\\alpha$", 
                          marker="^", c="r", **params)
    pluto.plot_mean_track(r22[filt][~ha][x_r], r22[filt][~ha][y_r], label="subgiants low $\\alpha$", 
                          marker="v", c="r", **params)

    x_l = x.upper().replace("O", "$\\alpha$")
    y_l = y.upper().replace("O", "$\\alpha$")
    plt.xlabel(x_l)
    plt.ylabel(y_l)
    
    pluto.legend_outside()
```

```python
def compare_means_hi_lo_panel(x, y, alpha=0.2, s=1, xlim=None, **kwargs):
    ha = v21["high_alpha"]
    c1 = pluto.COLORS[0]
    c2 = pluto.COLORS[3]
    
    params = {
        "xlim": xlim,
        "plot_points": False,
        "min_count": 3,
        "shade_width": True
    }
    pluto.plot_median_track(v21[ha][x], v21[ha][y], label="V+21 high $\\alpha$", 
                            ls="--", c=c1, **params)
    pluto.plot_median_track(v21[~ha][x], v21[~ha][y], label="V+21 low $\\alpha$", 
                            ls="-", c=c1, **params)

    x_r = aah.convert_name(x)
    y_r = aah.convert_name(y)
    
    filt = ~np.isnan(r22[x_r])
    filt &= ~np.isnan(r22[y_r])
    
    ha = r22["high_alpha"]
    pluto.plot_median_track(r22[filt][ha][x_r], r22[filt][ha][y_r], label="subgiants high $\\alpha$", 
                            ls="--", c=c2, **params)
    pluto.plot_median_track(r22[filt][~ha][x_r], r22[filt][~ha][y_r], label="subgiants low $\\alpha$", 
                            ls="-", c=c2, **params)

    x_l = x.upper().replace("O", "$\\alpha$")
    y_l = y.upper().replace("O", "$\\alpha$")
    plt.xlabel(x_l)
    plt.ylabel(y_l)
```

```python
def means_hist(x, y, xlim, ylim=None, xlabel=None, ylabel=None, legend=False):
    fig, axs = plt.subplots(2, 2, sharex="col", sharey="row",
                           gridspec_kw={'width_ratios': [3, 1],
                                       "height_ratios": [1, 3]})
    
    c1 = pluto.COLORS[0]
    c2 = pluto.COLORS[3]

    plt.subplots_adjust(hspace=0, wspace=0)
    plt.sca(axs[1][0])
    compare_means_hi_lo_panel(x, y, xlim=xlim)
    plt.ylim(ylim)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    if legend:
        pluto.legend_outside((1.4,1))

    hist_params = {
        "range": xlim,
        "bins": 20,
    }

    plt.sca(axs[0][0])
    ha = v21["high_alpha"]
    pluto.smooth_hist(v21[ha][x], color=c1, **hist_params)
    pluto.smooth_hist(v21[~ha][x], color=c1, ls="--", **hist_params)
    
    ha = r22["high_alpha"]
    x_r = aah.convert_name(x)
    y_r = aah.convert_name(y)
    pluto.smooth_hist(r22[ha][x_r], color=c2, **hist_params)
    pluto.smooth_hist(r22[~ha][x_r], color=c2, ls="--", **hist_params)

    plt.xlim(xlim)
    plt.yticks([])

    plt.sca(axs[1][1])
    hist_params["range"] = ylim
    hist_params["orientation"] = "horizontal"
    ha = v21["high_alpha"]
    pluto.smooth_hist(v21[ha][y], color=c1, **hist_params)
    pluto.smooth_hist(v21[~ha][y], color=c1, ls="--", **hist_params)
    
    ha = r22["high_alpha"]

    pluto.smooth_hist(r22[ha][y_r], color=c2, **hist_params)
    pluto.smooth_hist(r22[~ha][y_r], color=c2, ls="--", **hist_params)
    plt.xticks([])

    fig.delaxes(axs[0][1])
```
