```python
import vice
import numpy as np
import matplotlib.pyplot as plt
from plotting import fig_saver
```

```python
sf = fig_saver("figures/")
```

```python
example = vice.singlezone(name = "example")
from vice.yields.presets import JW20 
example.run(np.linspace(0, 10, 1001), overwrite=True)
example_out = vice.output("example")
# example_out.show("[fe/h]")
```

```python
def infall_rate(t):
    delta_t = 1
    delta_M = 5*10**9
    t0 = 5
    if (t0 <= t and t < t0 + delta_t):
        return 9.1 + delta_M/delta_t/10**9
    else:
        return 9.1

example = vice.singlezone(name="gas-burst", func=infall_rate, mode="ifr", dt=0.001)
example.run(np.linspace(0, 10, 1001), overwrite=True)
gas_burst = vice.output("gas-burst")

# gas_burst.show("[o/fe]-[fe/h]")
plt.plot(gas_burst.history["[fe/h]"], gas_burst.history["[o/fe]"])
plt.xlabel("[fe/h]")
plt.ylabel("[o/fe]")
sf("sz_ofefeh_gasburst")
#gas_burst.show("dn/d[o/fe]")
#gas_burst.show("sfr")
```

```python
gas_burst.show("ifr")
```

```python
def tau_star(t):
    delta_t = 1
    t0 = 5
    if (t0 <= t and t < t0 + delta_t):
        return 1
    else:
        return 2

example = vice.singlezone(name="eff-burst", tau_star=tau_star, mode="ifr", dt=0.001, schmidt=True, smoothing=0)
example.run(np.linspace(0, 10, 1001), overwrite=True)
eff_burst = vice.output("eff-burst")

#eff_burst.show("[o/fe]-[fe/h]")
plt.plot(eff_burst.history["[fe/h]"], eff_burst.history["[o/fe]"])
plt.xlabel("[fe/h]")
plt.ylabel("[o/fe]")
sf("sz_ofefeh_effburst")
#eff_burst.show("dn/d[o/fe]")
#eff_burst.show("sfr")
```

```python
def tau_star(t):
    return 2 + 0.2*np.cos(2*np.pi*t/2)

example = vice.singlezone(name="sin_hist", tau_star=tau_star, mode="ifr", dt=0.001, schmidt=True, smoothing=0, bins=np.arange(-3,1,0.01))
example.run(np.linspace(0, 10, 1001), overwrite=True)
sin_hist = vice.output("sin_hist")

sin_hist.show("[o/fe]-[fe/h]")
sin_hist.show("dn/d[o/fe]", xlim=(0,0.6))
sin_hist.show("sfr")
```
